<h1><center>Une régression linéaire simple</center></h1>

## Présentation des données

Nous utilisons le jeu de données `cars` disponible avec le langage R.

```r
head(cars)
```

## Régression linéaire simple

```r
model <- lm(dist~speed, data=cars)
cat(
  "Équation de la droite de régression :\n",
  "dist =", model$coefficients[1], "+", model$coefficients[2], "* speed\n"
)
```

```r
plot(cars$speed, cars$dist, xlab="Speed", ylab="Distance")
abline(model, col="green")
```
